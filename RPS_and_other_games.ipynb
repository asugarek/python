```python
#Provides the factorial from a given number

number = int(input("Please enter a number: "))
factor = 1

if number < 0:
    print ("Sorry, factorials do not exist for negative numbers.")
elif number == 0:
    print ("The factorial of 0 is 1.")
else:
    for i in range (1, number + 1):
        factor = factor * i
    print ("The factorial of ",number,"is ", factor)
```

```python
#Provides the Fibonaci series from a given integer

def fibonaci (x):
    if x <= 1:
        return x
    else:
        return (fibonaci (x - 1) + fibonaci(x - 2))

x = int(input ("Please enter a positive interger greater than 1: "))

fibseries = []

for y in range (0,x):
    fibseries.append(fibonaci(y))
    
print (fibseries)
```
